```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
%matplotlib inline
```

```python
dados = pd.read_csv("https://raw.githubusercontent.com/schunckbr/unilasalle/main/gorjetas.csv", sep=",")
sns.relplot(data=dados, x="sexo", y="gorjeta", hue="gorjeta")
plt.title('Densidade do total de gorjetas por sexo!')
```

```python
sns.relplot(data=dados, x="quantidade", y="gorjeta", hue="gorjeta")
plt.title('Densidade do total de gorjetas pela quantidade de pessoas na mesa.')
```

```python
sns.relplot(data=dados, x="dia", y="gorjeta", hue="dia")
plt.title('Densidade do total de gorjetas por dia da semana.')
```
